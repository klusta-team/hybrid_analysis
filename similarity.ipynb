# Computing similarity measures of hybrid clusters

```python
import numpy as np
import matplotlib.pyplot as plt
import kwiklib as kl
import klustaviewa as kv
import klustaviewa.stats as ks
%matplotlib inline
```

### Random data

```python
nspikes = 100000
nfeatures = 3*32
```

```python
spiketimes = np.cumsum(np.random.randint(size=nspikes, low=0, high=10000)).astype(np.int64)
spiketimes
```

This is the spike clusters, assigning one cluster to each spike.

```python
spike_clusters = np.random.randint(size=nspikes, low=0, high=100)
spike_clusters
```

This is the sorted list of clusters.

```python
clusters = np.unique(spike_clusters)
clusters
```

Let's create the features and masks arrays.

```python
features = np.random.randn(nspikes, nfeatures)
masks = np.random.rand(nspikes, nfeatures)
```

### Hybrids

```python
hybrid_indices = slice(None, None, 50)  # ::50
hybrid_spikes = spiketimes[hybrid_indices]
hybrid_spikes
```

```python
hybrid_amplitudes = np.random.uniform(size=len(hybrid_spikes), low=np.sqrt(2.)/2, high=np.sqrt(2.))
hybrid_amplitudes
```

Artificial split of the hybrid cluster into two clusters.

```python
cluh1 = hybrid_amplitudes<1.
cluh2 = hybrid_amplitudes>1.
```

```python
hybrid1 = hybrid_spikes[cluh1]
hybrid2 = hybrid_spikes[cluh2]
hybrid1.shape, hybrid2.shape
```

Now, we have the spike times of the two hybrid clusters.

### Similarity

Let's compute the global similarity matrix for all clusters.

```python
mat = ks.compute_correlations_approximation(features, spike_clusters, masks)
# here, mat is a dictionary (cluidx_i, cluidx_j) ==> value.
# below, we convert it to a regular (nclusters, nclusters) matrix.
mat = ks.get_similarity_matrix(mat)
```

```python
plt.imshow(mat);
```

Now, we're going to create an alternative `spike_clusters` array with two special clusters 1000 and 1001. These two clusters correspond to the hybrid spikes with amplitude `<1`, or `>1`.

```python
spike_clusters_h = spike_clusters.copy()
spike_clusters_h[hybrid_indices][cluh1] = 1000
spike_clusters_h[hybrid_indices][cluh2] = 1001
spike_clusters_h
```

```python
math = ks.compute_correlations_approximation(features, spike_clusters_h, masks)
math = ks.get_similarity_matrix(math)
```

```python
math[-2,:]  # similarity between hybrid cluster 1 and all other clusters
math[:,-1]  # similarity between hybrid cluster 2 and all other clusters
math[-2,-1]  # similarity between the two hybrid clusters
```
